```python
# Importing Libraries

import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.__version__
```

```python
# Importing dataset as Pandas dataframes

df_devices = pd.read_csv('LoRaWAN_devices.csv')
#devices = sns.load_dataset(df_devices)
df_devices.shape
```

```python
# Checking data visiually

df_devices.head(3)
```

```python
# Plotting Variety of curves

sns.relplot(x = 'Tx_Power', y = 'Cost', hue = 'Vendor' , col = 'Device_Category' , data = df_devices);
```

```python
sns.relplot(x = 'Tx_Power', y = 'Range', hue = 'Device_Category' , data = df_devices);
```

### Kernel Density Estimation

```python
# Kernel Density estimation -- Univariate

sns.kdeplot(df_devices.Tx_Power, 
            shade = True, 
            bw_adjust = 1,
            color = 'green'
           );
```

```python
# Kernel Density enstimation -- Bivariate

sns.kdeplot(x = 'Tx_Power' , y = 'RSSI', 
            shade = True, 
            data = df_devices,
            cbar = True
           );
```

```python
# Kernel Density enstimation -- Bivariate

sns.kdeplot(x = 'Rating' , y = 'Range', 
            shade = True,  
            color = 'red',
            data = df_devices,
            cbar = True
           );
```

### Histogram 

```python
# Horizontal histogram
sns.histplot(y = 'Rating', data = df_devices, 
             binwidth = 10, binrange = (50, 100),
             hue = 'Device_Category'
            ); 
```

```python
# Probability distribution

sns.histplot(x = 'Range', data = df_devices, 
             kde = True,
             stat = 'probability',
             binwidth = 2, binrange = (0, 15),
             hue = 'Device_Category'
            ); 
```

```python
# Polygon representation

sns.histplot(x = 'Cost', data = df_devices, 
             fill = True,
             element = 'poly',
             binrange = (0, 500),
             hue = 'Vendor'
            ); 
```

```python
# Bivariate Histogram

sns.histplot(x = 'Cost', y = 'Rating', 
             data = df_devices, 
             cbar = True,
             hue = 'Device_Category'
            ); 
```
